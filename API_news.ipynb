```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# List of topics you want to fetch data for
topics = ['battery','battery cell','car battery','electric car','portable power']  # Add more topics as needed

# Initialize an empty list to store articles
all_articles = []

# Define the start date
start_date = datetime.strptime('2024-02-18', '%Y-%m-%d')

# Iterate over each date for 5 days
for i in range(5):
    # Calculate the current date in the iteration
    current_date = start_date + timedelta(days=i)
    formatted_date = current_date.strftime('%Y-%m-%d')
    
    # Iterate over each topic
    for topic in topics:
        url = ('https://newsapi.org/v2/everything?'
               f'q={topic}&'
               f'from={formatted_date}&'
               'sortBy=popularity&'
               'pageSize=100&'
               'apiKey=4fea412182cf472aaadd75e7c81e0ac9')

        response = requests.get(url)
        data = response.json()
        
        # Check if articles are present in the response
        if 'articles' in data:
            articles = data['articles']
            # Add topic and date as columns in each article
            for article in articles:
                article['topic'] = topic
                article['date'] = formatted_date
            all_articles.extend(articles)

# Create a DataFrame from the list of articles
df = pd.DataFrame(all_articles)

# Display DataFrame
print(df.head())
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

news_content = df['title']
news_content
```

```python
import re

news_content_cleaned = [re.sub(r'[^a-zA-Z\s]', '', article) for article in news_content]

news_content_cleaned[:3]
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
news_token = []

for article in news_content_cleaned:
    # Tokenize the article
    tokens = word_tokenize(article.lower())
    # Remove stopwords
    tokens = [token for token in tokens if token not in stop_words]
    # Lemmatize the tokens
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    news_token.append(tokens)

# Print the tokenized news data
for i, article_tokens in enumerate(news_token):
  
    print(f"Article {i+1} Tokens:", article_tokens)
```

```python
news_token
```

```python
from collections import Counter

# Count the frequency of each token in the entire list
token_counts = Counter(token for sublist in news_token for token in sublist)

# Define a threshold frequency for rare tokens
threshold = 2  # Adjust as needed

# Identify rare tokens
rare_tokens = {token for token, count in token_counts.items() if count < threshold}

# Remove rare tokens from all sublists and trim each sublist to a maximum length of 5
news_token_processed = [[token for token in sublist if token not in rare_tokens][:5] for sublist in news_token]

# Print the processed list
print(news_token_processed)
```

```python
from collections import Counter
import itertools

# Flatten the list of lists to create a single list of all tokens
all_tokens = list(itertools.chain.from_iterable(news_token))

# Count the frequency of each token
token_counts = Counter(all_tokens)

# Select the top 20 most frequent tokens
top_tokens = [token for token, _ in token_counts.most_common(20)]

# Filter each sublist to include only the tokens from the top 20 list
filtered_news_token = [[token for token in sublist if token in top_tokens] for sublist in news_token]

# Remove sublists with no tokens after filtering
filtered_news_token = [sublist for sublist in filtered_news_token if sublist]

# Print the filtered list of lists
print(filtered_news_token)
```

```python
# Write cleaned_text to a CSV file
with open('multibattery_token_filter.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(filtered_news_token)
```
