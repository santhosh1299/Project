```python
import requests
import pandas as pd

# List of dates you want to fetch data for
dates = ['2024-01-23', '2024-01-24', '2024-01-25', '2024-01-26', '2024-01-27',
         '2024-01-28', '2024-01-29', '2024-01-30', '2024-01-31', '2024-02-01',
         '2024-02-02', '2024-02-03', '2024-02-04', '2024-02-05', '2024-02-06',
         '2024-02-07', '2024-02-08', '2024-02-09', '2024-02-10', '2024-02-11',
         '2024-02-12', '2024-02-13', '2024-02-14', '2024-02-15', '2024-02-16',
         '2024-02-17', '2024-02-18', '2024-02-19', '2024-02-20', '2024-02-21']  # Add more dates as needed

# Initialize an empty list to store articles
all_articles = []

# Iterate over dates
for date in dates:
    url = ('https://newsapi.org/v2/everything?'
           f'q=battery&'
           f'from={date}&'
           'sortBy=popularity&'
           'pageSize=100&'
           'apiKey=4fea412182cf472aaadd75e7c81e0ac9')

    response = requests.get(url)
    data = response.json()
    
    # Check if articles are present in the response
    if 'articles' in data:
        articles = data['articles']
        all_articles.extend(articles)

# Create a DataFrame from the list of articles
df = pd.DataFrame(all_articles)

# Display DataFrame
print(df.head())
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

news_content = df['title']
news_content
```

```python
df1 = df[df["title"].str.contains("battery", case=False)]
```

```python
news_content = df1['title']
news_content
```

```python
import re

news_content_cleaned = [re.sub(r'[^a-zA-Z\s]', '', article) for article in news_content]

news_content_cleaned[:3]
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
news_token = []

for article in news_content_cleaned:
    # Tokenize the article
    tokens = word_tokenize(article.lower())
    # Remove stopwords
    tokens = [token for token in tokens if token not in stop_words]
    # Lemmatize the tokens
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    news_token.append(tokens)

# Print the tokenized news data
for i, article_tokens in enumerate(news_token):
  
    print(f"Article {i+1} Tokens:", article_tokens)
```

```python
# Write cleaned_text to a CSV file
with open('news_title_batteryfilter.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(news_token)
```
